# Loads modules.
module load bedtools/2.25.0

# Defines variables.
dir=/scratch/Users/dara6367/hal_testing
inputBED=Human.hg38-tfit-3kb-ext-MUMERGE.bed
species_name_from=Homo_sapiens
species_assembly_from=hg38

# Creates temporary directories.
mkdir -p ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}

# Loops through each entry in the input BED file.
for bidir in `cut -f4 ${dir}/${inputBED}`
do
        # Feeds each separate BED file to HAL to obtain individual output BED files.
        /Users/dara6367/repos/hal/bin/halLiftover \
        /Shares/txpnevol/cactusGenomicAlignments/241-mammalian-2020v2.hal \
        ${species_name_from} \
        ${dir}/hg38-individual-bidirs/${bidir}.${inputBED} \
        ${species_name_to} \
        ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}/${bidir}.tmp1.bed

        # Sorts the individual output BED files.
        sort \
        -k 1,1 -k2,2n \
        ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}/${bidir}.tmp1.bed \
        > ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}/${bidir}.tmp2.bed

        # Merges all the entries that are within 1000 bp of each other.
        mergeBed \
        -d 1000 \
        -i ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}/${bidir}.tmp2.bed \
        > ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}/${bidir}.tmp3.bed

        # Adds 5th column with the length of each interval.
        awk -v OFS='\t' '{print $1, $2, $3, $3-$2}' \
        ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}/${bidir}.tmp3.bed \
        > ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}/${bidir}.tmp4.bed

        # Sorts intervals by their length (4th column) in descending order.
        sort -t $'\t' -k4,4rn \
        ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}/${bidir}.tmp4.bed \
        > ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}/${bidir}.tmp5.bed

        # Keeps only the biggest merged region per BED file.
        head -1 ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}/${bidir}.tmp5.bed \
        > ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}/${bidir}.tmp6.bed

        # Replaces the 4th width column with the entry name.
        awk -v OFS='\t' '{print $1, $2, $3}' \
        ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}/${bidir}.tmp6.bed \
        > ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}/${bidir}-${species_assembly_from}-${species_assembly_to}.bed
        sed -i "s/$/\t${bidir}/g" ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}/${bidir}-${species_assembly_from}-${species_assembly_to}.bed

        # Removes intermediary files.
        rm ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}/${bidir}.tmp1.bed
        rm ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}/${bidir}.tmp2.bed
        rm ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}/${bidir}.tmp3.bed
        rm ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}/${bidir}.tmp4.bed
        rm ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}/${bidir}.tmp5.bed
        rm ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}/${bidir}.tmp6.bed
done

# Combines all individual merged BED files onto a single file.
printf '%s\0' ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}/*-${species_assembly_from}-${species_assembly_to}.bed | \
xargs -r0 cat -- > ${dir}/hal-bidirs-${species_assembly_from}-${species_assembly_to}.bed

# Renames the chr to actual chr names.
for oldID in `cut -f1 ${dir}/chromAlias/table-${species_assembly_to}.txt`
        do
        newID=`grep ${oldID} ${dir}/chromAlias/table-${species_assembly_to}.txt | cut -f2`
        sed -i "s/${oldID}/${newID}/g" ${dir}/hal-bidirs-${species_assembly_from}-${species_assembly_to}.bed
done

# Removes intermediary files.
rm -R ${dir}/bidirs-${species_assembly_from}-${species_assembly_to}/

````r
# Makes data frame / dictionary with species, assembly, assembly size, number of TSS, and color.
speciesList <- c("Human","Chimp","Bonobo","Gorilla","Orangutan","Gibbon",
                 "Rhesus","Baboon","SquirrelMonkey","OwlMonkey")

# Creates color scheme data frame that matches species with their color.
speciesDict <- data.frame(matrix(ncol = 4, nrow = length(speciesList)))
colnames(speciesDict) <- c("species","assembly","colorCode","notTSSBidirsThres")
rownames(speciesDict) <- speciesList
speciesDict$species <- speciesList
speciesDict$assembly <- c("hg38","panTro6","panPan2","Susie3","ponAbe3","nomLeu3",
                          "rheMac8","papAnu4","saiBol1","Anan_2.0")
speciesDict$colorCode <- append("black",viridis(9))
speciesDict$notTSSBidirsThres <- c(3633,4688,5180,4149,3600,8818,7211,5160,8813,5175)
````

````r
# Defines variables.
species <- "Chimp"
baseDir <- "/scratch/Users/dara6367/hal_testing/"

conditionsTableFile_human <- paste0(baseDir,"featureCounts/conditionsTable.PRO-Nutlin-Human-hg38.csv")
featurecounts_genes_human <- paste0(baseDir,"featureCounts/PRO-Nutlin-Human.genes-hg38.featureCounts.RData")
featurecounts_bidirs_human <- paste0(baseDir,"featureCounts/PRO-Nutlin-Human.hg38.bidirs.notTSS.standard10.featureCounts.RData")

conditionsTableFile_other <- paste0(baseDir,"featureCounts/conditionsTable.PRO-Nutlin-",species,"-",speciesDict[species,"assembly"],".csv")
featurecounts_genes_other <- paste0(baseDir,"featureCounts/PRO-Nutlin-",species,".genes-hg38-",speciesDict[species,"assembly"],".featureCounts.RData")
featurecounts_bidirs_other <- paste0(baseDir,"featureCounts/PRO-Nutlin-",species,".",speciesDict[species,"assembly"],".bidirs.notTSS.standard10.featureCounts.RData")

############################################################################################################
# Reads in count data, conditions tables, and loci (genes and bidirs) lenghts.
# Adds a pseudocount of 1 to avoid INF values when doing log2 on TPM values.
############################################################################################################

# Human genes.
load(featurecounts_genes_human)
counts_genes_human <- as.data.frame(counts$counts)
counts_genes_human <- counts_genes_human + 1
conditionsTable_human <- read.csv(conditionsTableFile_human, header = TRUE)
colnames(counts_genes_human) <- conditionsTable_human$sample
rownames(conditionsTable_human) <- conditionsTable_human$sample
lengths_genes_human <- read.table(paste0(baseDir,"hal-output/hg38.ncbiRefSeq.oneEntry.saf"), 
                          header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")
lengths_genes_human$length <- lengths_genes_human$V4 - lengths_genes_human$V3
colnames(lengths_genes_human) <- c("gene","chr","start","end","strand","length")

# Human bidirs.
load(featurecounts_bidirs_human)
counts_bidirs_human <- as.data.frame(counts$counts)
counts_bidirs_human <- counts_bidirs_human + 1
colnames(counts_bidirs_human) <- conditionsTable_human$sample
lengths_bidirs_human <- read.table(paste0(baseDir,"hal-output/standard10-bidirs.notTSS.hg38.saf"), 
                          header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")
lengths_bidirs_human$length <- lengths_bidirs_human$V4 - lengths_bidirs_human$V3
colnames(lengths_bidirs_human) <- c("bidir","chr","start","end","dot","length")

# Other genes
load(featurecounts_genes_other)
counts_genes_other <- as.data.frame(counts$counts)
counts_genes_other <- counts_genes_other + 1
conditionsTable_other <- read.csv(conditionsTableFile_other, header = TRUE)
colnames(counts_genes_other) <- conditionsTable_other$sample
rownames(conditionsTable_other) <- conditionsTable_other$sample
lengths_genes_other <- read.table(paste0(baseDir,"hal-output/hal-genes-hg38-",speciesDict[species,"assembly"],".saf"), 
                          header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")
lengths_genes_other$length <- lengths_genes_other$V4 - lengths_genes_other$V3
colnames(lengths_genes_other) <- c("bidir","chr","start","end","strand","length")

# Other bidirs.
load(featurecounts_bidirs_other)
counts_bidirs_other <- as.data.frame(counts$counts)
counts_bidirs_other <- counts_bidirs_other + 1
colnames(counts_bidirs_other) <- conditionsTable_other$sample
lengths_bidirs_other <- read.table(paste0(baseDir,"hal-output/standard10-bidirs.notTSS.",speciesDict[species,"assembly"],".saf"), 
                          header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")
lengths_bidirs_other$length <- lengths_bidirs_other$V4 - lengths_bidirs_other$V3
colnames(lengths_bidirs_other) <- c("bidir","chr","start","end","dot","length")

############################################################################################################
# Removes DMSO columns, averages Nutlin columns, and adds the loci length as a new column to counts table.
############################################################################################################

# Human genes.
counts_genes_human <- counts_genes_human[, grepl("Nutlin", colnames(counts_genes_human))]
counts_genes_human$average_human <- (counts_genes_human[,1] + counts_genes_human[,2]) / 2
counts_genes_human$length_human <- lengths_genes_human$length/1000
counts_genes_human <- counts_genes_human[, !(grepl("Nutlin", colnames(counts_genes_human)))]

# Human bidirs.
counts_bidirs_human <- counts_bidirs_human[, grepl("Nutlin", colnames(counts_bidirs_human))]
counts_bidirs_human$average_human <- (counts_bidirs_human[,1] + counts_bidirs_human[,2]) / 2
counts_bidirs_human$length_human <- lengths_bidirs_human$length/1000
counts_bidirs_human <- counts_bidirs_human[, !(grepl("Nutlin", colnames(counts_bidirs_human)))]

# Other genes.
counts_genes_other <- counts_genes_other[, grepl("Nutlin", colnames(counts_genes_other))]
counts_genes_other$average_other <- (counts_genes_other[,1] + counts_genes_other[,2]) / 2
counts_genes_other$length_other <- lengths_genes_other$length/1000
counts_genes_other <- counts_genes_other[, !(grepl("Nutlin", colnames(counts_genes_other)))]

# Other bidirs.
counts_bidirs_other <- counts_bidirs_other[, grepl("Nutlin", colnames(counts_bidirs_other))]
counts_bidirs_other$average_other <- (counts_bidirs_other[,1] + counts_bidirs_other[,2]) / 2
counts_bidirs_other$length_other <- lengths_bidirs_other$length/1000
counts_bidirs_other <- counts_bidirs_other[, !(grepl("Nutlin", colnames(counts_bidirs_other)))]

############################################################################################################
# Calculates TPM per counts table.
############################################################################################################

# Human genes.
counts_genes_human$rpk_human <- counts_genes_human$average_human / counts_genes_human$length_human
perMillionScalingFactor <- (sum(counts_genes_human$rpk_human))/1000000
counts_genes_human$tpm_human_genes <- counts_genes_human$rpk_human / perMillionScalingFactor
counts_genes_human <- counts_genes_human[,colnames(counts_genes_human) == "tpm_human_genes", drop = FALSE]

# Human bidirs.
counts_bidirs_human$rpk_human <- counts_bidirs_human$average_human / counts_bidirs_human$length_human
perMillionScalingFactor <- (sum(counts_bidirs_human$rpk_human))/1000000
counts_bidirs_human$tpm_human_bidirs <- counts_bidirs_human$rpk_human / perMillionScalingFactor
counts_bidirs_human <- counts_bidirs_human[,colnames(counts_bidirs_human) == "tpm_human_bidirs", drop = FALSE]

# Other genes.
counts_genes_other$rpk_other <- counts_genes_other$average_other / counts_genes_other$length_other
perMillionScalingFactor <- (sum(counts_genes_other$rpk_other))/1000000
counts_genes_other$tpm_other_genes <- counts_genes_other$rpk_other / perMillionScalingFactor
counts_genes_other <- counts_genes_other[,colnames(counts_genes_other) == "tpm_other_genes", drop = FALSE]

# Other bidirs.
counts_bidirs_other$rpk_other <- counts_bidirs_other$average_other / counts_bidirs_other$length_other
perMillionScalingFactor <- (sum(counts_bidirs_other$rpk_other))/1000000
counts_bidirs_other$tpm_other_bidirs <- counts_bidirs_other$rpk_other / perMillionScalingFactor
counts_bidirs_other <- counts_bidirs_other[,colnames(counts_bidirs_other) == "tpm_other_bidirs", drop = FALSE]
````

````r
start.time <- Sys.time()

# Merges the counts tables, renames rows, removes redundant name column, and removes rows with NAs.
counts_genes_merged <- merge(counts_genes_human, counts_genes_other, by = "row.names", all = TRUE)
rownames(counts_genes_merged) <- counts_genes_merged$Row.names
counts_genes_merged <- counts_genes_merged[,!(colnames(counts_genes_merged) == "Row.names")]
counts_genes_merged <- counts_genes_merged[complete.cases(counts_genes_merged), ]

# Sorts bidirs by bidir ID in decreasing order, then concatenates the bidir TPM columns.
counts_bidirs_human <- counts_bidirs_human[order(rownames(counts_bidirs_human)),,drop = FALSE]
counts_bidirs_other <- counts_bidirs_other[order(rownames(counts_bidirs_other)),,drop = FALSE]
counts_bidirs_merged <- cbind(counts_bidirs_human, counts_bidirs_other)

end.time <- Sys.time()
time.taken <- end.time - start.time
time.taken
````
