```r
# Modifies the 'plotEnrichment' so that I can now specify the color of the enrichment line.

plotEnrichment_2 <- function(pathway, stats, otherColor, gseaParam = 1, ticksSize = 0.2) {
    rnk <- rank(-stats)
    ord <- order(rnk)
    statsAdj <- stats[ord]
    statsAdj <- sign(statsAdj) * (abs(statsAdj)^gseaParam)
    statsAdj <- statsAdj/max(abs(statsAdj))
    pathway <- unname(as.vector(na.omit(match(pathway, names(statsAdj)))))
    pathway <- sort(pathway)
    gseaRes <- calcGseaStat(statsAdj, selectedStats = pathway, 
        returnAllExtremes = TRUE)
    bottoms <- gseaRes$bottoms
    tops <- gseaRes$tops
    n <- length(statsAdj)
    xs <- as.vector(rbind(pathway - 1, pathway))
    ys <- as.vector(rbind(bottoms, tops))
    toPlot <- data.frame(x = c(0, xs, n + 1), y = c(0, ys, 0))
    diff <- (max(tops) - min(bottoms))/8
    x = y = NULL
    g <- ggplot(toPlot, aes(x = x, y = y)) + geom_point(color = otherColor, 
        size = 1) + geom_hline(yintercept = max(tops), colour = "red", 
        linetype = "dashed") + geom_hline(yintercept = min(bottoms), 
        colour = "red", linetype = "dashed") + geom_hline(yintercept = 0, 
        colour = "black") + geom_line(color = otherColor, size = 2) + theme_bw() + 
        suppressWarnings(geom_segment(data = data.frame(x = pathway), mapping = aes(x = x, 
            y = -diff/2, xend = x, yend = diff/2), size = ticksSize)) + 
        theme(panel.border = element_blank(), panel.grid.minor = element_blank()) + 
        labs(x = "rank", y = "enrichment score")
    g }
```

### Obtains DEG gene set and a ranked fold-change gene list per species

```r
start.time <- Sys.time()

# Runs DESeq2 in a loop, each time with a different species.
# Per species, it stores the DEG set, as well as fold-change for all the Standard-10 list.

contrast <- c("treatment","Nutlin","DMSO")
alphaValue <- 0.05

speciesList <- c("Human","Chimp","Bonobo","Gorilla","Orangutan", 
                 "Gibbon","Rhesus","Baboon","SquirrelMonkey","OwlMonkey")

for (species in speciesList) {
    # Keeps only the normalized counts from the chosen species libraries.
    counts_all_filt_2 <- counts_all_filt[, grepl(species, colnames(counts_all_filt))]
    conditionsTable_all_2 <- conditionsTable_all[grepl(species, rownames(conditionsTable_all)),]

    # Creates DESeq2Dataset object and load counts matrix.
    dds <- DESeqDataSetFromMatrix(countData = counts_all_filt_2, 
                                  colData = conditionsTable_all_2,
                                  design = ~ treatment)

    # Runs DESeq2 using the counts.
    dds <- suppressMessages(DESeq(dds))

    resultsUnshrunken <- results(dds, 
                                 contrast = contrast, 
                                 alpha = alphaValue)

    # Shrinks results using 'normal' mode.
    results <- suppressMessages(lfcShrink(dds, 
                         contrast = contrast, 
                         res = resultsUnshrunken, 
                         type = "normal"))

    # Substracts list of significantly differentially expressed genes per pvalue.
    resultsSig <- subset(results, padj < alphaValue)
    resultsSigPos <- subset(resultsSig, log2FoldChange > 0)
    resultsSigNeg <- subset(resultsSig, log2FoldChange < 0)
    assign(paste0("results_complete_", species), resultsSig)
    assign(paste0("results_complete_pos_", species), resultsSigPos)
    assign(paste0("results_complete_neg_", species), resultsSigNeg)

    # Obtains a GSEA-compatible (gmt-like) list of DEGs.
    
        # For all significant genes.
        geneSetHeader <- paste0("p53_RNA_", species)
        # Adds header to species-specific gmt list.
        p53_set <- c(geneSetHeader, rownames(resultsSig))
        write.table(t(p53_set), paste0("/Shares/txpnevol/daniel/gsea_tmp/gsea.",species,".tmp.gmt"), 
                    sep = "\t", quote = FALSE, row.names = FALSE, col.names = FALSE)
        p53_set_gmt <- scan(paste0("/Shares/txpnevol/daniel/gsea_tmp/gsea.",species,".tmp.gmt"), what = "", sep = "\n")
        # Transforms gene vector onto a list.
        p53_set_list <- strsplit(p53_set_gmt, "[[:space:]]+")
        # Sets the gene header as the list name.
        names(p53_set_list) <- sapply(p53_set_list, function(p53_set_gmt) p53_set_gmt[[1]])
        p53_set_list <- lapply(p53_set_list, function(p53_set_gmt) p53_set_gmt[-1])
        assign(paste0("p53_gmt_", species), p53_set_list)
                   
        # Only for significant genes with a positive log2FoldChange.                       
        geneSetHeader <- paste0("p53_RNA_pos_", species)
        # Adds header to species-specific gmt list.
        p53_set_pos <- c(geneSetHeader, rownames(resultsSigPos))
        write.table(t(p53_set_pos), paste0("/Shares/txpnevol/daniel/gsea_tmp/gsea.pos.",species,".tmp.gmt"), 
                    sep = "\t", quote = FALSE, row.names = FALSE, col.names = FALSE)
        p53_set_pos_gmt <- scan(paste0("/Shares/txpnevol/daniel/gsea_tmp/gsea.pos.",species,".tmp.gmt"), what = "", sep = "\n")
        # Transforms gene vector onto a list.
        p53_set_pos_list <- strsplit(p53_set_pos_gmt, "[[:space:]]+")
        # Sets the gene header as the list name.
        names(p53_set_pos_list) <- sapply(p53_set_pos_list, function(p53_set_pos_gmt) p53_set_pos_gmt[[1]])
        p53_set_pos_list <- lapply(p53_set_pos_list, function(p53_set_pos_gmt) p53_set_pos_gmt[-1])
        assign(paste0("p53_pos_gmt_", species), p53_set_pos_list)

        # Only for significant genes with a negative log2FoldChange.                       
        geneSetHeader <- paste0("p53_RNA_neg_", species)
        # Adds header to species-specific gmt list.
        p53_set_neg <- c(geneSetHeader, rownames(resultsSigNeg))
        write.table(t(p53_set_neg), paste0("/Shares/txpnevol/daniel/gsea_tmp/gsea.neg.",species,".tmp.gmt"), 
                    sep = "\t", quote = FALSE, row.names = FALSE, col.names = FALSE)
        p53_set_neg_gmt <- scan(paste0("/Shares/txpnevol/daniel/gsea_tmp/gsea.neg.",species,".tmp.gmt"), what = "", sep = "\n")
        # Transforms gene vector onto a list.
        p53_set_neg_list <- strsplit(p53_set_neg_gmt, "[[:space:]]+")
        # Sets the gene header as the list name.
        names(p53_set_neg_list) <- sapply(p53_set_neg_list, function(p53_set_neg_gmt) p53_set_neg_gmt[[1]])
        p53_set_neg_list <- lapply(p53_set_neg_list, function(p53_set_neg_gmt) p53_set_neg_gmt[-1])
        assign(paste0("p53_neg_gmt_", species), p53_set_neg_list)
                                   
    # Obtains a GSEA-compatible list of all Standard-10 genes ranked in descending order by log2foldchange.
        # Filter out gene entries whose log2FoldChange has an NA value.
        filtresults <- results[!is.na(results$log2FoldChange),]
        # Orders the gene list by log2FoldChange in decreasing order.
        filtresults <- filtresults[order(filtresults$log2FoldChange, decreasing = TRUE),]
        rankedGenes <- filtresults$log2FoldChange
        names(rankedGenes) <- rownames(filtresults)
        assign(paste0("rankedGenes_", species), rankedGenes)
}

end.time <- Sys.time()
time.taken <- end.time - start.time
time.taken
```

### Using all significant genes per species (i.e. positive and negative log2FoldChanges)

```r
start.time <- Sys.time()

# Obtains the GSEA NES looping through each species ranked genes and each species p53 gene set.

speciesList <- c("Human","Chimp","Bonobo","Gorilla","Orangutan", 
                 "Gibbon","Rhesus","Baboon","SquirrelMonkey","OwlMonkey")

# Initializes data frame that will contain the GSEA results.
gsea_ES_results <- data.frame(matrix(ncol = 11, nrow = 10))
colnames(gsea_ES_results) <- c("species","p53_Human","p53_Chimp","p53_Bonobo","p53_Gorilla","p53_Orangutan", 
                            "p53_Gibbon","p53_Rhesus","p53_Baboon","p53_SquirrelMonkey","p53_OwlMonkey")
gsea_ES_results$species <- speciesList
gsea_NES_results <- data.frame(matrix(ncol = 11, nrow = 10))
colnames(gsea_NES_results) <- c("species","p53_Human","p53_Chimp","p53_Bonobo","p53_Gorilla","p53_Orangutan", 
                            "p53_Gibbon","p53_Rhesus","p53_Baboon","p53_SquirrelMonkey","p53_OwlMonkey")
gsea_NES_results$species <- speciesList

# Outer loop takes each species ranked gene list.
for(species_rankedGenes in speciesList) {
    # Inner loop takes each species p53 GMT gene set.
    for(species_p53_gmt in speciesList) {
        # Runs GSEA on the selected species, and stores the NES in the master GSEA results table.
        fgseaRes <- suppressWarnings(fgsea(pathways = eval(as.symbol(paste0("p53_gmt_", species_p53_gmt))), 
                          stats = eval(as.symbol(paste0("rankedGenes_", species_rankedGenes))),
                          nperm = 5000))
        gsea_ES_results[gsea_ES_results$species == species_rankedGenes, 
                    colnames(gsea_ES_results) == paste0("p53_",species_p53_gmt)] <- as.numeric(fgseaRes[,"ES"])
        gsea_NES_results[gsea_NES_results$species == species_rankedGenes, 
                    colnames(gsea_NES_results) == paste0("p53_",species_p53_gmt)] <- as.numeric(fgseaRes[,"NES"])
    }
}

gsea_ES_results
gsea_NES_results

end.time <- Sys.time()
time.taken <- end.time - start.time
time.taken
```

### Obtains DEG gene set and a ranked fold-change gene list per species

```r
start.time <- Sys.time()

# Runs DESeq2 in a loop, each time with a different species.
# Per species, it stores the DEG set, as well as fold-change for all the Standard-10 list.

contrast <- c("treatment","Nutlin","DMSO")
alphaValue <- 0.05

speciesList <- c("Human","Chimp","Bonobo","Gorilla","Orangutan", 
                 "Gibbon","Rhesus","Baboon","SquirrelMonkey","OwlMonkey")

for (species in speciesList) {
    # Keeps only the normalized counts from the chosen species libraries.
    counts_all_filt_2 <- counts_all_filt[, grepl(species, colnames(counts_all_filt))]
    conditionsTable_all_2 <- conditionsTable_all[grepl(species, rownames(conditionsTable_all)),]

    # Creates DESeq2Dataset object and load counts matrix.
    dds <- DESeqDataSetFromMatrix(countData = counts_all_filt_2, 
                                  colData = conditionsTable_all_2,
                                  design = ~ treatment)

    # Runs DESeq2 using the counts.
    dds <- suppressMessages(DESeq(dds))

    resultsUnshrunken <- results(dds, 
                                 contrast = contrast, 
                                 alpha = alphaValue)

    # Shrinks results using 'normal' mode.
    results <- suppressMessages(lfcShrink(dds, 
                         contrast = contrast, 
                         res = resultsUnshrunken, 
                         type = "normal"))

    # Substracts list of significantly differentially expressed genes per pvalue.
    resultsSig <- subset(results, padj < alphaValue)
    assign(paste0("results_complete_", species), resultsSig)

    # Obtains a GSEA-compatible (gmt-like) list of DEGs.
        geneSetHeader <- paste0("p53_RNA_", species)
        # Adds header to species-specific gmt list.
        p53_set <- c(geneSetHeader, rownames(resultsSig))
        write.table(t(p53_set), paste0("/Shares/txpnevol/daniel/gsea_tmp/gsea.",species,".tmp.gmt"), 
                    sep = "\t", quote = FALSE, row.names = FALSE, col.names = FALSE)
        p53_set_gmt <- scan(paste0("/Shares/txpnevol/daniel/gsea_tmp/gsea.",species,".tmp.gmt"), what = "", sep = "\n")
        # Transforms gene vector onto a list.
        p53_set_list <- strsplit(p53_set_gmt, "[[:space:]]+")
        # Sets the gene header as the list name.
        names(p53_set_list) <- sapply(p53_set_list, function(p53_set_gmt) p53_set_gmt[[1]])
        p53_set_list <- lapply(p53_set_list, function(p53_set_gmt) p53_set_gmt[-1])
        assign(paste0("p53_gmt_", species), p53_set_list)

    # Obtains a GSEA-compatible list of all Standard-10 genes ranked in descending order by log2foldchange.
        # Filter out gene entries whose log2FoldChange has an NA value.
        filtresults <- results[!is.na(results$log2FoldChange),]
        # Orders the gene list by log2FoldChange in decreasing order.
        filtresults <- filtresults[order(filtresults$log2FoldChange, decreasing = TRUE),]
        rankedGenes <- filtresults$log2FoldChange
        names(rankedGenes) <- rownames(filtresults)
        assign(paste0("rankedGenes_", species), rankedGenes)
}

end.time <- Sys.time()
time.taken <- end.time - start.time
time.taken
```

### Obtains the GSEA NES score for all combinations of each species gene list and p53-gene set

```r
start.time <- Sys.time()

# Obtains the GSEA NES looping through each species ranked genes and each species p53 gene set.

speciesList <- c("Human","Chimp","Bonobo","Gorilla","Orangutan", 
                 "Gibbon","Rhesus","Baboon","SquirrelMonkey","OwlMonkey")

# Initializes data frame that will contain the GSEA results.
gsea_results <- data.frame(matrix(ncol = 11, nrow = 10))
colnames(gsea_results) <- c("species","p53_Human","p53_Chimp","p53_Bonobo","p53_Gorilla","p53_Orangutan", 
                            "p53_Gibbon","p53_Rhesus","p53_Baboon","p53_SquirrelMonkey","p53_OwlMonkey")
gsea_results$species <- speciesList

# Outer loop takes each species ranked gene list.
for(species_rankedGenes in speciesList) {
    # Inner loop takes each species p53 GMT gene set.
    for(species_p53_gmt in speciesList) {
        # Runs GSEA on the selected species, and stores the NES in the master GSEA results table.
        fgseaRes <- suppressWarnings(fgsea(pathways = eval(as.symbol(paste0("p53_gmt_", species_p53_gmt))), 
                          stats = eval(as.symbol(paste0("rankedGenes_", species_rankedGenes))),
                          nperm = 5000))
        gsea_results[gsea_results$species == species_rankedGenes, 
                     colnames(gsea_results) == paste0("p53_",species_p53_gmt)] <- as.numeric(fgseaRes[,"NES"])
    }
}

gsea_results

end.time <- Sys.time()
time.taken <- end.time - start.time
time.taken
```
