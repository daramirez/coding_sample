```r
##########################################################
               # Loading RNA-seq data
##########################################################

inDir <- "/Shares/txpnevol/daniel/DESeq2files"

# Loads the RNA-seq condition table.
conditionsTable_RNA_all <- read.table(paste0(inDir,"/conditionsTable_all_RNA.tsv"), 
                                      row.names = 1, header = TRUE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")

# Loads the raw RNA-seq counts table.
counts_RNA_all <- read.table(paste0(inDir,"/counts_raw_RNA_all_filt.tsv"), 
                     row.names = 1, header = TRUE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")
colnames(counts_RNA_all) <- gsub("\\.", "-", names(counts_RNA_all))

# Loads the list of panPrimate p53 genes from RNA-seq.
names_sig_RNA_all <- read.table(paste0(inDir,"/names_sig_RNA_all.tsv"), 
                     header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")
names_sig_RNA_all <- names_sig_RNA_all$V1

p53_names_downstream <- read.table(paste0(inDir,"/p53_names_downstream.tsv"), 
                     header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")
p53_names_downstream <- p53_names_downstream$V1

# Removes all genes from Standard-10 that are not in the panPrimate p53 gene set.
counts_RNA_all_filt <- counts_RNA_all[rownames(counts_RNA_all) %in% names_sig_RNA_all,]

# Removes the gene entries that have an NA in any species.
counts_RNA_all_filt <- counts_RNA_all_filt[complete.cases(counts_RNA_all_filt),]

# Reads in names_sig* files per primate.
names_sig_RNA_Human <- read.table(paste0(inDir,"/names_sig_RNA_Human.tsv"), 
                                  header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")$V1
names_sig_RNA_Chimp <- read.table(paste0(inDir,"/names_sig_RNA_Chimp.tsv"), 
                                  header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")$V1
names_sig_RNA_Bonobo <- read.table(paste0(inDir,"/names_sig_RNA_Bonobo.tsv"), 
                                  header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")$V1
names_sig_RNA_Gorilla <- read.table(paste0(inDir,"/names_sig_RNA_Gorilla.tsv"), 
                                  header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")$V1
names_sig_RNA_Orangutan <- read.table(paste0(inDir,"/names_sig_RNA_Orangutan.tsv"), 
                                  header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")$V1
names_sig_RNA_Gibbon <- read.table(paste0(inDir,"/names_sig_RNA_Gibbon.tsv"), 
                                  header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")$V1
names_sig_RNA_Rhesus <- read.table(paste0(inDir,"/names_sig_RNA_Rhesus.tsv"), 
                                  header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")$V1
names_sig_RNA_Baboon <- read.table(paste0(inDir,"/names_sig_RNA_Baboon.tsv"), 
                                  header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")$V1
names_sig_RNA_SquirrelMonkey <- read.table(paste0(inDir,"/names_sig_RNA_SquirrelMonkey.tsv"), 
                                  header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")$V1
names_sig_RNA_OwlMonkey <- read.table(paste0(inDir,"/names_sig_RNA_OwlMonkey.tsv"), 
                                  header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")$V1

##########################################################
               # Loading PRO-seq data
##########################################################

# Loads the PRO-seq condition table.
conditionsTable_PRO_all <- read.table(paste0(inDir,"/conditionsTable_all_PRO.tsv"), 
                                      row.names = 1, header = TRUE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")

# Loads the raw PRO-seq counts table.
counts_PRO_all <- read.table(paste0(inDir,"/counts_raw_PRO_all_filt.tsv"), 
                     row.names = 1, header = TRUE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")
colnames(counts_PRO_all) <- gsub("\\.", "-", names(counts_PRO_all))

# Loads the list of panPrimate p53 genes from PRO-seq and primary gene set.
names_sig_PRO_all <- read.table(paste0(inDir,"/names_sig_PRO_all.tsv"), 
                     header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")
names_sig_PRO_all <- names_sig_PRO_all$V1

p53_names_primary <- read.table(paste0(inDir,"/p53_names_primary.tsv"), 
                     header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")
p53_names_primary <- p53_names_primary$V1

# Removes all genes from Standard-10 that are not in the panPrimate p53 gene set.
counts_PRO_all_filt <- counts_PRO_all[rownames(counts_PRO_all) %in% names_sig_PRO_all,]

# Removes the gene entries that have an NA in any species.
counts_PRO_all_filt <- counts_PRO_all_filt[complete.cases(counts_PRO_all_filt),]

# Reads in names_sig* files per primate.
names_sig_PRO_Human <- read.table(paste0(inDir,"/names_sig_PRO_Human.tsv"), 
                                  header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")$V1
names_sig_PRO_Chimp <- NA

names_sig_PRO_Bonobo <- read.table(paste0(inDir,"/names_sig_PRO_Bonobo.tsv"), 
                                  header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")$V1
names_sig_PRO_Gorilla <- read.table(paste0(inDir,"/names_sig_PRO_Gorilla.tsv"), 
                                  header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")$V1
names_sig_PRO_Orangutan <- read.table(paste0(inDir,"/names_sig_PRO_Orangutan.tsv"), 
                                  header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")$V1
names_sig_PRO_Gibbon <- read.table(paste0(inDir,"/names_sig_PRO_Gibbon.tsv"), 
                                  header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")$V1
names_sig_PRO_Rhesus <- read.table(paste0(inDir,"/names_sig_PRO_Rhesus.tsv"), 
                                  header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")$V1
names_sig_PRO_Baboon <- read.table(paste0(inDir,"/names_sig_PRO_Baboon.tsv"), 
                                  header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")$V1
names_sig_PRO_SquirrelMonkey <- read.table(paste0(inDir,"/names_sig_PRO_SquirrelMonkey.tsv"), 
                                  header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")$V1
names_sig_PRO_OwlMonkey <- read.table(paste0(inDir,"/names_sig_PRO_OwlMonkey.tsv"), 
                                  header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")$V1
```

```r
# Determines in how many species each gene is statistically significant.

speciesList <- c("Human","Chimp","Bonobo","Gorilla","Orangutan", 
                 "Gibbon","Rhesus","Baboon","SquirrelMonkey","OwlMonkey")
colNames <- append(speciesList, c("Conservation","Motif"))

# Initializes data frame that will contain the significance binary labels and the number of instances.
genes_sig_all_RNA <- data.frame(matrix(FALSE, ncol = length(colNames), nrow = dim(counts_RNA_all_filt)[1]))
colnames(genes_sig_all_RNA) <- colNames
rownames(genes_sig_all_RNA) <- rownames(counts_RNA_all_filt)
genes_sig_all_PRO <- data.frame(matrix(FALSE, ncol = length(colNames), nrow = dim(counts_PRO_all_filt)[1]))
colnames(genes_sig_all_PRO) <- colNames
rownames(genes_sig_all_PRO) <- rownames(counts_PRO_all_filt)

# Labels each gene significance per species.
for(species in speciesList) {    
    binaryLabels <- rownames(genes_sig_all_RNA) %in% eval(as.symbol(paste0("names_sig_RNA_", species)))
    genes_sig_all_RNA[,species] <- binaryLabels }
for(species in speciesList) {    
    binaryLabels <- rownames(genes_sig_all_PRO) %in% eval(as.symbol(paste0("names_sig_PRO_", species)))
    genes_sig_all_PRO[,species] <- binaryLabels }

# Sums the number of species in which each gene is significant.
for(gene in rownames(genes_sig_all_RNA)) {
    genes_sig_all_RNA[gene,"Conservation"] <- length(genes_sig_all_RNA[gene,][genes_sig_all_RNA[gene,] == TRUE]) }
for(gene in rownames(genes_sig_all_PRO)) {
    genes_sig_all_PRO[gene,"Conservation"] <- length(genes_sig_all_PRO[gene,][genes_sig_all_PRO[gene,] == TRUE]) }

# Determines if each gene has an associated TP53 motif in its TSS.
Human_TSS_TP53 <- read.table("/Shares/txpnevol/daniel/FIMO-assemblies-TP53/hg38-standard10-TSS-3kb-FIMO-TP53_M06704_1.1e-5.txt", 
                         header = TRUE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")
Human_TSS_TP53_names <- unique(Human_TSS_TP53$sequence_name)
genes_sig_all_RNA$Motif <- rownames(genes_sig_all_RNA) %in% Human_TSS_TP53_names
genes_sig_all_PRO$Motif <- rownames(genes_sig_all_PRO) %in% Human_TSS_TP53_names
```

```r
##########################################
            # PRO-seq Heatmap
##########################################

# Plot binary heatmap given a pan primate p53 gene list.

# Initializes data frame that will hold the binary notation for each p53-responsive gene per species
speciesList <- c("Human","Chimp","Bonobo","Gorilla","Orangutan", 
                 "Gibbon","Rhesus","Baboon","SquirrelMonkey","OwlMonkey")

pan_p53_binary <- data.frame(matrix(0, ncol = length(speciesList), nrow = length(names_sig_PRO_all)))
colnames(pan_p53_binary) <- speciesList
rownames(pan_p53_binary) <- names_sig_PRO_all

# Fills data frame, adding '1' if the gene is significant in each species.
for(gene in names_sig_PRO_all) {
    if (gene %in% names_sig_PRO_Human) { pan_p53_binary[gene,"Human"] <- 1 }
    if (gene %in% names_sig_PRO_Chimp) { pan_p53_binary[gene,"Chimp"] <- 1 }
    if (gene %in% names_sig_PRO_Bonobo) { pan_p53_binary[gene,"Bonobo"] <- 1 }
    if (gene %in% names_sig_PRO_Gorilla) { pan_p53_binary[gene,"Gorilla"] <- 1 }
    if (gene %in% names_sig_PRO_Orangutan) { pan_p53_binary[gene,"Orangutan"] <- 1 }
    if (gene %in% names_sig_PRO_Gibbon) { pan_p53_binary[gene,"Gibbon"] <- 1 }
    if (gene %in% names_sig_PRO_Rhesus) { pan_p53_binary[gene,"Rhesus"] <- 1 }
    if (gene %in% names_sig_PRO_Baboon) { pan_p53_binary[gene,"Baboon"] <- 1 }
    if (gene %in% names_sig_PRO_SquirrelMonkey) { pan_p53_binary[gene,"SquirrelMonkey"] <- 1 }
    if (gene %in% names_sig_PRO_OwlMonkey) { pan_p53_binary[gene,"OwlMonkey"] <- 1 }
}

# Determines the number of species conservation for heatmap color labels.
genes_sig_all_PRO_filt <- genes_sig_all_PRO[rownames(genes_sig_all_PRO) %in% rownames(pan_p53_binary),]
genes_sig_all_PRO_filt <- genes_sig_all_PRO_filt[,c("Conservation","Motif"), drop = FALSE]
rowNames <- rownames(genes_sig_all_PRO_filt)
genes_sig_all_PRO_filt <- data.frame(Conservation = as.character(genes_sig_all_PRO_filt$Conservation),
                                     Motif = as.character(genes_sig_all_PRO_filt$Motif))
rownames(genes_sig_all_PRO_filt) <- rowNames

# Sorts genes by the number of species where each is conserved.
genes_sig_all_PRO_filt_sorted <- genes_sig_all_PRO_filt[order(as.numeric(genes_sig_all_PRO_filt$Conservation)),,drop = FALSE]
pan_p53_binary_sorted <- pan_p53_binary[rownames(genes_sig_all_PRO_filt_sorted),]

# Resorts the binary data frame by species.
pan_p53_binary <- pan_p53_binary[with(pan_p53_binary, order(-Human,-Chimp,-Bonobo,-Gorilla,-Orangutan,
                                                              -Gibbon,-Rhesus,-Baboon,-SquirrelMonkey,-OwlMonkey)), ]

# Transforms the binary data frame onto a matrix
pan_p53_binary_sorted <- t(as.matrix(pan_p53_binary_sorted))
pan_p53_binary <- t(as.matrix(pan_p53_binary))

annotationColors <- list(Conservation = c(inferno(length(unique(genes_sig_all_PRO_filt$Conservation)))), 
                         Motif = c("#482878FF","#6DCD59FF"))
names(annotationColors$Conservation) <- c(as.character(sort(as.numeric(unique(genes_sig_all_PRO_filt$Conservation)))))
names(annotationColors$Motif) <- c(FALSE,TRUE)

# Plots heatmap.
options(repr.plot.width = 16, repr.plot.height = 6)
setHook("grid.newpage", function() pushViewport(viewport(x = 1,y = 1,width = 0.9, 
                                                         height = 0.9, name = "vp", 
                                                         just = c("right","top"))), action = "prepend")
pheatmap(mat = pan_p53_binary,
         color = c("white","black"),
         border_color = NA,
         show_colnames = FALSE,
         labels_row = speciesList,
         cluster_rows = FALSE, 
         cluster_cols = FALSE,
         main = paste0("PRO-seq\nn = ",dim(pan_p53_binary)[2]),
         fontsize = 18,
         legend = FALSE,
         scale = "none",
         annotation_col = genes_sig_all_PRO_filt,
         annotation_colors = annotationColors)
setHook("grid.newpage", NULL, "replace")
grid.text("DEG", x = 0.375, y = -0.02, gp = gpar(fontsize = 30))
```

```r
# Heatmap of p53 and TF categories.

# Adds a new column with the sum of columns to obtain conservation bin.
pan_p53_binary_t <- t(pan_p53_binary)
pan_p53_binary_t <- cbind(pan_p53_binary_t,as.data.frame(rowSums(pan_p53_binary_t, dims = 1)))
colnames(pan_p53_binary_t) <- c(speciesList,"conservation")

# Resorts data frame in descending conservation order, and removes the conservation column.
pan_p53_binary_t <- pan_p53_binary_t[ ,-which(names(pan_p53_binary_t) %in% speciesList), drop = FALSE]
pan_p53_binary_t$isCore_Andrysik2017 <- FALSE
# pan_p53_binary_t$isTF_cisBP <- FALSE
pan_p53_binary_t$isTF_Vaquerizas2009 <- FALSE
pan_p53_binary_t$isTF_Lambert2018 <- FALSE
pan_p53_binary_t <- pan_p53_binary_t[order(pan_p53_binary_t$conservatio, decreasing = TRUE),]
pan_p53_binary_t <- pan_p53_binary_t[pan_p53_binary_t$conservation > 0, ]
pan_p53_binary_t <- pan_p53_binary_t[ ,-which(names(pan_p53_binary_t) %in% "conservation"), drop = FALSE]

# Loads the categories.
p53coreList_Andrysik2017 <- read.table("/Shares/txpnevol/daniel/panSpeciesGeneAnnotation/misc/p53coreList_Andrysik2017.tsv",stringsAsFactors = FALSE)
colnames(p53coreList_Andrysik2017) <- "genes"
humanTFlist_Vaquerizas2009 <- read.table("/Shares/txpnevol/daniel/panSpeciesGeneAnnotation/misc/humanTFlist_Vaquerizas2009.tsv",stringsAsFactors = FALSE)
colnames(humanTFlist_Vaquerizas2009) <- c("ID","genes")
humanTFlist_Vaquerizas2009 <- humanTFlist_Vaquerizas2009[,"genes",drop = FALSE]
humanTFlist_Lambert2018 <- read.table("/Shares/txpnevol/daniel/panSpeciesGeneAnnotation/misc/humanTFlist_Lambert2018.tsv",stringsAsFactors = FALSE)
colnames(humanTFlist_Lambert2018) <- "genes"
# humanTFlist_cisBP <- read.table("/Shares/txpnevol/daniel/panSpeciesGeneAnnotation/misc/humanTFlist_cisBP.tsv",stringsAsFactors = FALSE)
# colnames(humanTFlist_cisBP) <- "genes"

# Fills in matrix with TRUE if that gene is in each category.
pan_p53_binary_t$isCore_Andrysik2017 <- rownames(pan_p53_binary_t) %in% p53coreList_Andrysik2017$genes
#pan_p53_binary_t$isTF_cisBP <- rownames(pan_p53_binary_t) %in% humanTFlist_cisBP$genes
pan_p53_binary_t$isTF_Vaquerizas2009 <- rownames(pan_p53_binary_t) %in% humanTFlist_Vaquerizas2009$genes
pan_p53_binary_t$isTF_Lambert2018 <- rownames(pan_p53_binary_t) %in% humanTFlist_Lambert2018$genes

# Removes "TP53 motif at TSS" column from color track.
genes_sig_all_RNA_filt_2 <- genes_sig_all_RNA_filt[,"Conservation",drop = FALSE]

# Plots heatmap.
options(repr.plot.width = 16, repr.plot.height = 5)
pheatmap(mat = t(pan_p53_binary_t * 1),
         color = c("white","black"),
         border_color = NA,
         show_colnames = FALSE,
         cluster_rows = FALSE, 
         cluster_cols = FALSE,
         main = paste0("Primary\nn = ",dim(pan_p53_binary_t)[1]),
         fontsize = 18,
         legend = FALSE,
         scale = "none",
         annotation_col = genes_sig_all_RNA_filt_2,
         annotation_colors = annotationColors)
```
