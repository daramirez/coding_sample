````r
# Defines variables.
species <- "Human"
bidirs <- "notTSS"
motif <- "MA0106_3_TP53"
slope_bin <- 500

# Makes data frame / dictionary with species, assembly, assembly size, number of TSS, and color.
speciesList <- c("Human","Chimp","Bonobo","Gorilla","Orangutan","Gibbon",
                 "Rhesus","Baboon","SquirrelMonkey","OwlMonkey")

# Creates color scheme data frame that matches species with their color.
speciesDict <- data.frame(matrix(ncol = 3, nrow = length(speciesList)))
colnames(speciesDict) <- c("species","assembly","colorCode")
rownames(speciesDict) <- speciesList
speciesDict$species <- speciesList
speciesDict$assembly <- c("hg38","panTro6","panPan2","Susie3","ponAbe3","nomLeu3",
                          "rheMac8","papAnu4","saiBol1","Anan_2.0")
speciesDict$colorCode <- append("black",viridis(9))

# Loads the TFEA data.
ranked_file <- read.table(paste0("/scratch/Users/dara6367/PRO-seq_interspecies-nutlin/tfea/standard10/",species,"/",bidirs,"/temp_files/ranked_file.bed.edited"), 
                        header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")
colnames(ranked_file) <- c("rank","bidir")
AUC <- read.table(paste0("/scratch/Users/dara6367/PRO-seq_interspecies-nutlin/tfea/standard10/",species,"/",bidirs,"/temp_files/withAUC",motif,"__dis_andmore.csv.edited"), 
                        header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")
colnames(AUC) <- c("rank2","score")
merged <- cbind(ranked_file,AUC)

# Normalizes score so that it goes from 0 to 1.
merged$fraction <- merged$score/max(merged$score)

# Normalizes rank and score so that they goes from 0 to 1.
merged$score_norm <- merged$score/max(merged$score)
merged$rank_norm <- merged$rank/max(merged$rank)

# Initializes data frame that will contain the slopes in bins.
slopes <- data.frame(matrix(ncol = 3 ,nrow = length(seq(from = 0, to = dim(merged)[1]-slope_bin, by = slope_bin))))
colnames(slopes) <- c("rank_norm","score_norm","slope")
slopes$rank_norm <- seq(from = 0, to = dim(merged)[1]-slope_bin, by = slope_bin)
rownames(slopes) <- seq(from = 0, to = dim(merged)[1]-slope_bin, by = slope_bin)

# # Collapses fraction measurements onto defined bins to reduce noise.
for(i in seq(from = 0, to = dim(slopes)[1]-2, by = 1)) {
    slopes[i,"score_norm"] <- mean(merged[seq(from = rownames(slopes[i+1,]), 
                                         to = as.numeric(rownames(slopes[i+2,]))-1, 
                                         by = 1),"score_norm"]) }

# Removes last entries with a slope NA.
slopes <- slopes[complete.cases(slopes$score_norm),]

# Normalizes rank so that it goes from 0 to 1.
slopes$rank_norm <- slopes$rank_norm/max(slopes$rank_norm)

# Calculates 1st derivative of the curve line.
# 1st derivative/Slope = (Yb - Ya) / (Xb - Xa)
for(i in seq(from = 0, to = dim(slopes)[1], by = 1)) {
    Ya <- slopes[i+1,"score_norm"]
    Yb <- slopes[i+2,"score_norm"]
    Xa <- slopes[i+1,"rank_norm"]
    Xb <- slopes[i+2,"rank_norm"]
    slopes[i+1,"slope"] <- (Yb - Ya) / (Xb - Xa) }

# Removes last entries with a slope NA.
slopes <- slopes[complete.cases(slopes$slope),]

# Finds the rank where the slope first becomes 1.
for(i in seq(from = 1, to = dim(slopes)[1], by  = 1)) {
    slopeBefore <- slopes[i,"slope"]
    slopeAfter <- slopes[i+1,"slope"]
    # Stores the rank position of the first time slope crosses 1.
    # Stops searching once the conditions are met.
    if(slopeBefore > 1 & slopeAfter < 1) { 
        # Determines which of the two ranks is the closest to 1.
        if(which.min(abs(c(slopeBefore,slopeAfter) - 1)) == 1) {
        pickedSlope <- slopeBefore
        pickedSlopeRank <- slopes[slopes$slope == pickedSlope,][1,"rank_norm"]
        pickedSlopeScore <- slopes[slopes$slope == pickedSlope,][1,"score_norm"]
        break } 
        if(which.min(abs(c(slopeBefore,slopeAfter) - 1)) == 2) {
        pickedSlope <- slopeAfter
        pickedSlopeRank <- slopes[slopes$slope == pickedSlope,][1,"rank_norm"]
        pickedSlopeScore <- slopes[slopes$slope == pickedSlope,][1,"score_norm"]
        break } } }

# Adds new column denoting until what rank to fill in with color in plots.
slopes$fill <- slopes$rank_norm <= pickedSlopeRank

# Defines threshold on ranked bidirs to be printed out.
bidirThreshold <- which.min(abs(merged$rank_norm - pickedSlopeRank))
bidirTotal <- dim(merged)[1]
bidirPercent <- format(round((bidirThreshold / bidirTotal)*100, 2), nsmall = 2)

# Stores curve and slope plots.
plot_curve <- ggplot(data = merged, mapping = aes(x = rank, y = fraction)) + 
    geom_line(lwd = 1.5, size = 1, color = speciesDict[species,"colorCode"]) +
    theme_classic(base_size = 20) +
    theme(legend.position = "none", 
        plot.title = element_text(color = "black", hjust = 0.5),
        axis.text.y = element_text(color = "black"),
        axis.text.x = element_text(color = "black"),
        axis.title.x = element_text(color = "black", hjust = 0.5)) +
    geom_abline(linetype = "solid", lwd = 0.5,intercept = 0, slope = max(merged$fraction) / max(merged$rank), color = "black") +
    labs(title = "Not binned\nCumulative score curve",
       x = "Ranked bidirectionals",
       y = "Cumulative score") +
    scale_x_continuous(limits = c(0,dim(merged)[1]), expand = c(0, 0)) +
    scale_y_continuous(limits = c(0,1), expand = c(0, 0))

plot_smooth_curve <- ggplot(data = slopes, mapping = aes(x = rank_norm, y = score_norm)) + 
    geom_line(lwd = 1.5, size = 1, color = speciesDict[species,"colorCode"]) +
    geom_area(aes(x = ifelse(rank_norm < pickedSlopeRank, rank_norm, pickedSlopeRank), y = score_norm), fill = speciesDict[species,"colorCode"]) +
    theme_classic(base_size = 20) +
    theme(legend.position = "none", 
        plot.title = element_text(color = "black", hjust = 0.5),
        axis.text.y = element_text(color = "black"),
        axis.text.x = element_text(color = "black"),
        axis.title.x = element_text(color = "black", hjust = 0.5)) +
    geom_abline(linetype = "solid", lwd = 0.5, intercept = 0, slope = 1, color = "black") +
    geom_abline(linetype = "solid", lwd = 1, intercept = pickedSlopeScore-pickedSlopeRank, slope = 1, color = "red") +
    labs(title = paste0("Bin size = ",slope_bin,"\nCumulative score curve"),
       x = "Normalized ranking",
       y = "Cumulative score") +
    scale_x_continuous(limits = c(0,1), expand = c(0, 0)) +
    scale_y_continuous(limits = c(0,1), expand = c(0, 0))

plot_smooth_slope <- ggplot(data = slopes, mapping = aes(x = rank_norm, y = slope)) + 
    geom_line(lwd = 1, size = 1, color = speciesDict[species,"colorCode"]) +
    geom_area(aes(x = ifelse(rank_norm < pickedSlopeRank, rank_norm, pickedSlopeRank), y = slope), fill = speciesDict[species,"colorCode"]) +
    theme_classic(base_size = 20) +
    theme(legend.position = "none", 
        plot.title = element_text(color = "black", hjust = 0.5),
        axis.text.y = element_text(color = "black"),
        axis.text.x = element_text(color = "black"),
        axis.title.x = element_text(color = "black", hjust = 0.5)) +
    labs(title = paste0("Bin size = ",slope_bin,"\nSlope"),
       x = "Normalized ranking",
       y = "Slope") +
    geom_hline(yintercept = pickedSlope, linetype = "dashed", lwd = 0.5, color = "black") +
    scale_x_continuous(limits = c(0,1), expand = c(0, 0)) +
    scale_y_continuous(limits = c(0,max(slopes$slope)), expand = c(0, 0), labels = function(x) sprintf("%.2f", x))

# Plots all graphs in a grid.
options(repr.plot.width = 10, repr.plot.height = 16)
grid.arrange(plot_curve, plot_smooth_curve, plot_smooth_slope,
             ncol = 1, nrow = 3,
             top = textGrob(paste0(species,"  ",bidirs," bidirs\n",motif,"\nTotal bidirectionals = ",dim(merged)[1],"\nInduced bidirectionals = 0 to ",bidirThreshold," (",bidirPercent,"%)\n"), 
                            gp = gpar(fontsize = 25))) 
````

````r
# Makes data frame / dictionary with species, assembly, assembly size, number of TSS, and color.
speciesList <- c("Human","Chimp","Bonobo","Gorilla","Orangutan","Gibbon",
                 "Rhesus","Baboon","SquirrelMonkey","OwlMonkey")

# Creates color scheme data frame that matches species with their color.
speciesDict <- data.frame(matrix(ncol = 5, nrow = length(speciesList)))
colnames(speciesDict) <- c("species","assembly","colorCode","colorCodeMiddle","notTSSBidirsThres")
rownames(speciesDict) <- speciesList
speciesDict$species <- speciesList
speciesDict$assembly <- c("hg38","panTro6","panPan2","Susie3","ponAbe3","nomLeu3",
                          "rheMac8","papAnu4","saiBol1","Anan_2.0")
speciesDict$colorCode <- append("#000000FF",viridis(9))
speciesDict$notTSSBidirsThres <- c(3633,4688,5180,4149,3600,8818,7211,5160,8813,5175)

# Determines middle color between white and each primate's color.
for(species in rownames(speciesDict)){
    white_int <- as.vector(col2rgb("#FFFFFFFF"))
    species_int <- as.vector(col2rgb(speciesDict[species,"colorCode"]))
    newR <- floor((white_int[1]+species_int[1])/2)
    newG <- floor((white_int[2]+species_int[2])/2)
    newB <- floor((white_int[3]+species_int[3])/2)
    speciesDict[species,"colorCodeMiddle"] <- paste0("#",toupper(as.character(as.hexmode(newR))),toupper(as.character(as.hexmode(newG))),toupper(as.character(as.hexmode(newB))),"FF") }
````

````r
# Defines variables.
TFdb <- "JASPAR2022_CORE_vertebrates_non-redundant"
motif <- "MA0106_3_TP53"
binsN <- 100

for(species in speciesList) {
    # Loads table contain the chosen motif positions.
    motifPositions <- read.table(paste0("/scratch/Users/dara6367/leadingEdgeFimo/",species,"/",species,"-",speciesDict[species,"assembly"],"-",TFdb,"-bgtm-p1e-6-tfit-positions3kb/",motif,".relative3kb.positions"), 
                               header = FALSE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")
    colnames(motifPositions) <- "positions"

    # Creates data frame that will contain the binned instances of chosen motif around bidirs origins.
    motifPositionsBins <- data.frame(matrix(0, ncol = 2, nrow = binsN))
    colnames(motifPositionsBins) <- c("positionBins","counts")
    motifPositionsBins$positionBins <- seq(from = 3000/binsN, to = 3000, by = 3000/binsN)
    rownames(motifPositionsBins) <- seq(from = 3000/binsN, to = 3000, by = 3000/binsN)
    for(bin in seq(from = 3000/binsN, to = 3000, by = 3000/binsN)) {
        motifPositionsBins[motifPositionsBins$positionBins == bin,"counts"] <- sum(motifPositions$positions[motifPositions$positions %in% seq(from = bin-(3000/binsN-1), to = bin)] / 
                                                                                   motifPositions$positions[motifPositions$positions %in% seq(from = bin-(3000/binsN-1), to = bin)]) }

    # Loads table contain the chosen motif counts.
    motifCounts <- read.table(paste0("/scratch/Users/dara6367/leadingEdgeFimo/",species,"/TFs-",species,"-",speciesDict[species,"assembly"],"-",TFdb,"-bgtm-FIMO_p1e-6-intersect-tfit-bidirs.txt"), 
                              header = TRUE, sep = "\t", fill = TRUE, stringsAsFactors = FALSE, na.strings = "")

    # Assigns the highest value from all max values across species.
    #maxBinN <- max(motifPositionsBins$counts)
    maxBinN <- 31

    # Stores barcode.
    barcode <- ggplot(motifPositionsBins, aes_string("positionBins", 1, fill = "counts")) +
        geom_tile() +
        scale_fill_gradient2(low = "white", #"#FFFFFFFF"
                             mid = "bisque3", #speciesDict[species,"colorCodeMiddle"]
                             high = "midnightblue", #speciesDict[species,"colorCode"]
                             na.value = "black",
                             #midpoint = (max(motifPositionsBins$counts)-min(motifPositionsBins$counts))/2,
                             midpoint = 15,
                             name = NULL,
                             limits = c(0,maxBinN)) +
        theme_void() +
        labs(title = paste0(species,
                            "\nbidirs = ",speciesDict[species,"notTSSBidirsThres"],
                            "     bins = ",binsN,
                            "\ncore = ",motifCounts[motifCounts$motif == motif,"core"],
                            "     extended = ", motifCounts[motifCounts$motif == motif,"extended"],
                            "     ratio = ",signif(motifCounts[motifCounts$motif == motif,"core"]/motifCounts[motifCounts$motif == motif,"extended"],2))) +
        theme(axis.text.x = element_text(size = 10), 
              plot.title = element_text(size = 20, hjust = 0.5),
              legend.position = "none") +
        scale_x_continuous(breaks = c(0,1350,1500,1650,3000), labels = c("-1.5 kb","|","0","|","1.5 kb"))

    # Puts species name to the barcode variable name.
    assign(paste0("barcode_",species), barcode)
}
    
# Plots all barcodes in a grid.
options(repr.plot.width = 24, repr.plot.height = 9)
grid.arrange(barcode_Human, barcode_Chimp, barcode_Bonobo, barcode_Gorilla, barcode_Orangutan,
             barcode_Gibbon, barcode_Rhesus, barcode_Baboon, barcode_SquirrelMonkey, barcode_OwlMonkey,
             ncol = 4, nrow = 3,
             top = textGrob(paste0("Leading-edge induced bidirectionals\n",TFdb,"\n",motif,"    FIMO p1e-6\n"), 
                            gp = gpar(fontsize = 30)))
````
